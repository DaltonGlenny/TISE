Import libraries for plotting and numerical work

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import matplotlib.gridspec as gridspec
import math
import copy
```

Set up parameters for the calculation

```python
x0 = 0 #init height (m)
V0 = math.inf #init voltage
u0 = 1
E = 3.81

dx = 0.1 #timestep
xf = 5 #final time (s)

nsteps = int((xf-x0)/dx) + 1 
```

Setting up arrays

```python
u = np.zeros([nsteps])
x = np.linspace(x0, xf, nsteps)
V = np.zeros([nsteps])
psi = np.zeros([nsteps])
```

Main loop

```python
u[0] = u0
V[0] = V0
psi[0] = 0

for i in range(1, nsteps):
  V[i] = x[i]
#   V[i] = V[i - 1] + 2 * (E - V[i - 1]) * psi[i - 1] * dx
  psi[i] = psi[i - 1] + V[i - 1] * dx
```

Plot a comparison of the methods:

```python
plt.plot(x,V)
# plt.title('Trajectory of ball w/ gravity, w/ drag')
# plt.xlabel('Time /s')
# plt.ylabel('Height /m')
plt.show()
plt.plot(x,psi)
plt.show()
```
